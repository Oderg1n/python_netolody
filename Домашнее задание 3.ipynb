Дана переменная, в которой хранится словарь, содержащий геометки для каждого пользователя (пример структуры данных приведён ниже). Вам нужно написать программу, которая выведет на экран множество уникальных геометок всех пользователей.

```python
# Пример работы программы:

ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}
# Результат: {98, 35, 15, 213, 54, 119}
```

#### Задание 2
Написать код на Python в среде Jupyter Notebook для решения следующей задачи.

Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведён ниже). Вам нужно написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт'
]
```

```python
# Считаем количество слов в каждом запросе
word_counts = [len(query.split()) for query in queries]

# Подсчитываем частоту каждого количества слов вручную
total_queries = len(queries)
count_dist = {}

for count in word_counts:
    if count in count_dist:
        count_dist[count] += 1
    else:
        count_dist[count] = 1

# Выводим результаты в требуемом формате
for word_count in sorted(count_dist.keys()):
    occurrences = count_dist[word_count]
    percentage = (occurrences / total_queries) * 100
    print(f"Поисковых запросов, содержащих {word_count} слов(а): {percentage:.2f}%")
```
