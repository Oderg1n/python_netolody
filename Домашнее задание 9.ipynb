### Задание 1
функция, которая принимает на вход строку и проверяет, является ли она валидной 
Валидируем транспортный номер: 1 буква, 3 цифры, 2 буквы, 2-3 цифры
Если номер валиден - возвращаем отдельно номер и отдельно регион.

Пример: 

car_id = 'А222BС96’
Результат: Номер А222BС валиден. Регион: 96.

car_id = 'АБ22ВВ193’
Результат: Номер не валиден.

```python
import re

def validate_car_number(car_id):
    """
    Валидация транспортного номера автомобиля.
    Формат: 1 буква, 3 цифры, 2 буквы, 2-3 цифры
    """
    # Регулярное выражение для проверки номера
    # [А-Я] - одна заглавная русская буква
    # \d{3} - три цифры
    # [А-Я]{2} - две заглавные русские буквы
    # \d{2,3} - две или три цифры (регион)
    pattern = r'^([А-Я])(\d{3})([А-Я]{2})(\d{2,3})$'
    
    match = re.match(pattern, car_id)
    
    if match:
        letter1 = match.group(1)
        digits = match.group(2)
        letters2 = match.group(3)
        region = match.group(4)
        
        number = letter1 + digits + letters2
        
        return f"Номер {number} валиден. Регион: {region}."
    else:
        return "Номер не валиден."


car_id1 = 'А222ВС96'
print(validate_car_number(car_id1))

car_id2 = 'АБ22ВВ193'
print(validate_car_number(car_id2))
```

### Задание 2
Функция, удаляющая последовательные повторы слов из строки

```python
import re

def remove_consecutive_duplicates(text):
    """
    Удаляет все последовательные повторы слов из строки.
    """
    # \b(\w+)\b - захватываем слово в группу 1
    # (?:\s+\1\b)+ - ищем один или более повторов этого же слова через пробелы
    # \1 - обратная ссылка на первую группу (повторяющееся слово)
    pattern = r'\b(\w+)\b(?:\s+\1\b)+'
    
    result = re.sub(pattern, r'\1', text)
    
    return result

some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'

print(f'Результат: {remove_consecutive_duplicates(some_string)}')
```
