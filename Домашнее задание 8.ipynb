## Задание № 1
Код, который возвращает название валюты с максимальным значением курса с помощью сервиса: https://www.cbr-xml-daily.ru/daily_json.js

```python
import requests

url = 'https://www.cbr-xml-daily.ru/daily_json.js'
response = requests.get(url)
data = response.json()

max_value = 0
max_currency_name = ""

for currency_code, currency_info in data['Valute'].items():
    rate_per_unit = currency_info['Value'] / currency_info['Nominal']
    
    if rate_per_unit > max_value:
        max_value = rate_per_unit
        max_currency_name = currency_info['Name']

print(max_currency_name)
```

## Задание № 2
Добавьте в класс Rate параметр diff (со значениями True или False), который, в случае значения True в методах курсов валют (eur, usd и т. д.), будет возвращать не курс валюты, а изменение по сравнению с прошлым значением. Считайте, что значение diff=True параметр будет принимать только при возврате значения курса (format_=‘value’). При отображении всей информации о валюте (format_=‘full’) он не используется.

Примеры работы программы:

1.	r = Rate(diff=True)
r.usd()

Результат (зависит от текущего курса):
-0.013721

2.	r = Rate(diff=False)
r.usd()

Результат (зависит от текущего курса):
60.5748


```python
import requests

class Rate:
    def __init__(self, diff=False):
        self.format = 'value'
        self.diff = diff
        self.data = self._load_data()
    
    def _load_data(self):
        """Загрузка данных с сайта ЦБ РФ"""
        url = 'https://www.cbr-xml-daily.ru/daily_json.js'
        response = requests.get(url)
        return response.json()
    
    def show_current_format(self):
        return self.format
    
    def _get_currency(self, currency_code):
        """Получение информации о валюте по коду"""
        if currency_code not in self.data['Valute']:
            return None
        
        currency_info = self.data['Valute'][currency_code]
        
        if self.format == 'value':
            if self.diff:
                return currency_info['Value'] - currency_info['Previous']
            else:
                return currency_info['Value']
        elif self.format == 'full':
            return currency_info
    
    def usd(self):
        """Курс доллара США"""
        return self._get_currency('USD')
    
    def eur(self):
        """Курс евро"""
        return self._get_currency('EUR')
    
    def gbp(self):
        """Курс фунта стерлингов"""
        return self._get_currency('GBP')
    
    def jpy(self):
        """Курс японской йены"""
        return self._get_currency('JPY')
    
    def cny(self):
        """Курс китайского юаня"""
        return self._get_currency('CNY')

r1 = Rate(diff=True)
print(r1.usd())  

#r2 = Rate(diff=False)
#print(r2.usd())  

r3 = Rate()
print(r3.usd()) 
```
