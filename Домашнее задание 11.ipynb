### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

- оценка 2 и ниже — низкий рейтинг;
- оценка 4 и ниже — средний рейтинг;
- оценка 4.5 и 5 — высокий рейтинг.
- 
Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies_stats = pd.read_csv('movies_stats.csv')
movies_stats.tail()
```

```python
movies_stats.info()
```

```python
movies_stats['class'] = pd.cut(
    movies_stats['rating'], 
    bins=[0, 2, 4, 5], 
    labels=['низкий рейтинг', 'средний рейтинг', 'высокий рейтинг'],
    include_lowest=True
)
print(movies_stats[['rating', 'class']].head(10))
```

### Задание 2
гео-классификатор, который каждой строке сможет выставить географическую при надлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
‘Центр’: [‘москва’, ‘тула’, ‘ярославль’],
‘Северо-Запад’: [‘петербург’, ‘псков’, ‘мурманск’],
‘Дальний Восток’: [‘владивосток’, ‘сахалин’, ‘хабаровск’]
}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
keywords.info()
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
```

```python
d = {}

for region in geo_data:
    
    spisok_gorodov = geo_data[region]
    
    d.update({key:value for key,value in zip(spisok_gorodov,
                                [region for i in spisok_gorodov])}) 
```

```python
keywords['region'] = keywords['keyword'].map(d)
```

```python
keywords['region'] = keywords['region'].fillna('undefined')
keywords.head()
```
