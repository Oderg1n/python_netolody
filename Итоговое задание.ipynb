### Условия задачи
Проверка теории: **верно ли, что с ростом года выпуска фильма средний рейтинг становится ниже**.

В соответствии с алгоритмом:

1. В переменную years запишите список из всех годов с 1950-го по 2010-й. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска (год выпуска берём из столбца title).

Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
- для каждой строки пройдите по всем годам списка years;
- если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска;
- если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.

4. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’.
5. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
import pandas as pd
```

```python
movies_stats = pd.read_csv('movies_stats.csv')
movies_stats.head()
```

```python
movies_stats.info()
```

```python
# достаём года из названий и записываем в новую колонку года
movies_stats['year'] = movies_stats['title'].str.extract(r'\((\d{4})\)')
movies_stats['year'] = pd.to_numeric(movies_stats['year'])

print(movies_stats[['title', 'year']].head())
```

```python
# проверяю, не выпали ли какие-то фильмы, где неочевидно указан год
movies_stats.info()
movies_stats['year'].describe()
# выпало 7 строк....
```

```python
# ищем строки с нулл-значениями года
null_years = movies_stats[movies_stats['year'].isna()]
print(null_years[['title', 'year']])
```

```python
def production_year(title):
    '''
    Ищем год в названии в формате 19xx или 20xx
    
    '''
    for i in range(len(title) - 3):
        if (title[i:i+2] in ['19', '20'] and 
            title[i+2:i+4].isdigit()):
            return int(title[i:i+4])
    return 1900

# Применяем только к строкам с Нан
movies_stats.loc[movies_stats['year'].isna(), 'year'] = \
    movies_stats.loc[movies_stats['year'].isna(), 'title'].apply(production_year)

print(movies_stats['year'].isna().sum())
print(movies_stats[['title', 'year']].head())
```

```python
# теперь все строки в year не нулевые
movies_stats.info()
```

```python
# групируем года и сотрируем по убыванию
avg_rating_by_year = movies_stats.groupby('year')['rating'].mean().sort_values(ascending=False)
# print(avg_rating_by_year.head(10))
```

```python
year_ratings = movies_stats.groupby('year')['rating'].mean().reset_index()
year_ratings.columns = ['year', 'avg_rating']
year_ratings = year_ratings.sort_values('avg_rating', ascending=False)
print(year_ratings.head(20))

# ну, кажется, выдвинутая теория в задании верна :) грустно!
```
