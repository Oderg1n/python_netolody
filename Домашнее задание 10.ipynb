### Задание 2
посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
%pwd
```

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
transactions = pd.read_csv('transactions.csv')
transactions.head()
```

```python
transactions.info()
```

```python
power.info()
```

```python
power['country'].unique()
```

```python
power[ power['country'].str.contains('Lit', case=False) ]['country'].unique()
```

```python
result = power[
    (power['category'].isin([4, 12, 21])) & 
    (power['country'].isin(['Latvia', 'Lithuania', 'Estonia'])) &
    (power['year'].between(2005, 2010)) &
    (power['quantity'] >= 0)
]
print(result)
```

```python
total_quantity = result['quantity'].sum()
print(f"Общая сумма quantity: {total_quantity}")
```
